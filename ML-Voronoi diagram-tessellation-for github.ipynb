## Voronoi diagram (tessellation)
We have a finite number of points called **seeds** in a *metric space* (a vector space with a distance function). Then, we can **partition** the space into disjoint regions called **Voronoi cells**. Specifically, a point of the space belongs to a seed if the point is closest to the seed than to any other seeds. The closeness is measured by the distance function of the metric space.
<br>**Contents:**
- Defining the distance functions **Euclidean** and **Manhattan**.
- Defining a function to generate random seeds, and then a function to create the Voronoi diagram.
- Examples with the same seeds but with different distance functions.

```python
# importing the required modules
import numpy as np
from matplotlib import pyplot as plt
```

```python
# the function to compute the Euclidean distance between two vectors
def Euclidean(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2,axis=1))

# the function to compute the Manhattan distance between two vectors
def Manhattan(x1,x2):
    return np.sum(np.abs(x1-x2),axis=1)

# a function to generate random seeds
# given the number of seeds and the size of the space
def generate_seeds(no_seeds=15,size=(200,500)):
    rows,cols=size
    # random seeds in the space of the image
    seeds=np.random.rand(noSeeds,2)
    seeds[:,0]*=rows; seeds[:,1]*=cols
    return seeds

# a function to create the Voronoi diagram given the seeds,
# the size of the space, and the distance function in that space
def generate_voronoi(seeds,size=(200,500),distance=Euclidean):
    rows,cols=size
    image=np.zeros((rows,cols),dtype=np.int16)
    for row in range(rows):
        for col in range(cols):
            point=np.array([row,col])
            image[row,col]=np.argmin(distance(seeds,point))
    return image
```

```python
# size of the 2D space as rows and columns of the image
rows,cols=200,500
# number of seeds
noSeeds=13
# random seeds in the space of the image
seeds=generate_seeds(noSeeds,(rows,cols))
```

```python
#assigning points of the 2D space (pixesl) to the seeds
# with Euclidean distance
image=generate_voronoi(seeds,(rows,cols))
# diaplying the Voronoi diagram (tessellation) along the seeds
plt.figure(figsize=(10,5))
plt.imshow(image,cmap=plt.cm.hot,alpha=.5)
plt.scatter(seeds[:,1],seeds[:,0],marker='o',s=30,c='b')
plt.title('The Voronoi diagram with Euclidean distance.\n The seeds are shown with black dots')
plt.show()
```

```python
#assigning points of the 2D space (pixesl) to the seeds
# with Manhattan distance
image=generate_voronoi(seeds,(rows,cols),Manhattan)
# diaplying the Voronoi diagram (tessellation) along the seeds
plt.figure(figsize=(10,5))
plt.imshow(image,cmap=plt.cm.hot,alpha=.5)
plt.scatter(seeds[:,1],seeds[:,0],marker='o',s=30,c='b')
plt.title('The Voronoi diagram with Manhattan distance.\n The seeds are shown with black dots')
plt.show()
```
