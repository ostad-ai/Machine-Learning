### Probability: Expected value 

The **expected value** (also called the **expectation** or **mean**) is a fundamental concept in probability theory and statistics. It represents the long-run average value of a random variable if an experiment is repeated many times.
<hr>

For a **discrete random variable** $X$, the expected value $E[X]$ is defined as:
<br>$E[X]=∑_{x_i} x_i⋅P(X=x_i)$
<br>where $P(X=x_i)$ is the probability of the random variable $X$ takes the value $x_i$.
<br>Or we can define the expected value with the probability (mass) function $f(x)$ of $X$:
<br>$E[X]=∑_x x⋅f(x)$
<hr>

For a **continuous random variable** $X$, the expected value $E[X]$ is defined as:
<br>$E[X]=\int_{-\infty}^{\infty} x\cdot f(x)dx$
<br>where $f(x)$ is the probability (density) function (PDF) of $X$.

```python
# Import thr required modules
import numpy as np
from matplotlib import pyplot as plt
# Import the required function for integration
from scipy.integrate import quad
```

```python
# Example: discrete random variable
# X represents the outcome of rolling a fair six-sided die
# Define the outcomes and their probabilities
outcomes = [1, 2, 3, 4, 5, 6]
# Each outcome has an equal probability (here, 1/6)
probabilities = [1/len(outcomes)] * len(outcomes)  

# Calculate the expected value
expected_value = sum(x * p for x, p in zip(outcomes, probabilities))

print(f"Expected value of rolling a fair six-sided die: {expected_value}")
```

```python
# let's draw the probability (mass) function for example above:
plt.bar(outcomes,probabilities,.05,color='orange')
plt.scatter(outcomes,probabilities,s=50,c='k')
# The expected value is shown with red mark
plt.scatter(expected_value,0,s=200,c='red',marker='o')
plt.title('Probability (mass) function for the discrete example of a fair die')
plt.show()
```

<hr style="height:5px; background-color:green">

```python
# Example: continuous random variable
# Define the probability (density) function (for short, PDF)
def pdf(x):
    return 2. * x if 0 <= x <= 1 else 0.

# Define the integrand for the expected value: x * f(x)
def integrand(x):
    return x * pdf(x)

# Compute the expected value using numerical integration
expected_value, _ = quad(integrand, -np.inf, np.inf)

print(f"Expected value of the continuous random variable: {expected_value}")
```

```python
# let's draw the probability (density) function for example above:
xs=np.linspace(-1,2,100)
pdf_vec=np.vectorize(pdf)
ys=pdf_vec(xs)
plt.plot(xs,ys,'-')
# The expected value is shown with red mark
plt.scatter(expected_value,0,s=100,c='red',marker='o')
plt.title('Probability (density) function of the continuous example')
plt.show()
```
