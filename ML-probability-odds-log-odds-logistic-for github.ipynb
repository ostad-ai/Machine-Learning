### From probability to log-odds
**Log-odds** (also called **logit**) is the logarithm of the **odds** of an event occurring. For an event with probability $p$:
<div style="padding:0;margin-top:5px;">$\large logit(p)=log\text{-}odds(p)=log⁡(\frac{p}{1−p})$</div>

Where:
- $\frac{p}{1-p}$ is the **odds** (with range $[0,+\infty)$). 
    - Odds are the **ratio** of the probability of an event *occurring* to the probability of it *not occurring*, expressing how much more likely the event is to happen than not.
- The logarithm is typically **natural** log (base $e$).
- Log-odds transforms the probability from $[0,1]$ to $(-\infty,+\infty)$.
<hr>

**Hint:** The **logistic function** is the **inverse** of log-odds. Thus, from log-odds, we can get to the probability $p$, by logistic function $\sigma(\cdot)$:
<div style="padding:0;margin:5px"> $\large p=\sigma(log\text{-}odds)=1/(1+e^{−log\text{-}odds})$ </div>
Where the logistic function is: $\sigma(x)=1/(1+e^{-x})$
<hr>

Some other properties of log-odds:
- **Symmetry**: log-odds$(1-p)=-$log-odds$(p)$
- **Derivative**: $\frac{d}{dp}$log-odds$(p)=\frac{1}{p(1-p)}$
<hr>

In the following, we display three curves: **odds**, **log-odds**, and finally inverse of log-odds, which is the **logistic** function.

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
def demonstrate_log_odds():
    """Visualize the relationship between probability, odds, log-odds,
       and logistic function."""
    
    fig, axes = plt.subplots(3,1, figsize=(6,15))
    
    # Probability values
    p = np.linspace(0.001, 0.999, 1000)
    
    # 1. Odds
    odds = p / (1 - p)
    
    # 2. Log-odds
    log_odds = np.log(odds)
    
    # 3. Logistic(log_odds)
    p_from_logodds = 1 / (1 + np.exp(-log_odds))
    
    # Plot 1: Odds
    ax = axes[0]
    ax.plot(p, odds, 'r-', linewidth=2)
    ax.axhline(y=1, color='gray', linestyle='--', alpha=0.5)
    ax.axvline(x=0.5, color='gray', linestyle='--', alpha=0.5)
    ax.set_xlabel('p')
    ax.set_ylabel('Odds')
    ax.set_title('$Odds=p/(1-p)$')
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 20)
    
    # Plot 2: Log-odds
    ax = axes[1]
    ax.plot(p, log_odds, 'g-', linewidth=2)
    ax.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
    ax.axvline(x=0.5, color='gray', linestyle='--', alpha=0.5)
    ax.set_xlabel('p')
    ax.set_ylabel('Log-Odds')
    ax.set_title(r'Log-Odds$=log(p/(1-p))$')
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 1)
    ax.set_ylim(-5, 5)
    
    # Plot 3: Logistic function (inverse of log-odds)
    ax = axes[2]
    ax.plot(log_odds, p_from_logodds, 'c-', linewidth=2)
    ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)
    ax.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
    ax.set_xlabel('Log-Odds')
    ax.set_ylabel('p')
    ax.set_title(r'Logistic Function, from Log-Odds: $p = 1/(1 + e^{-Log-Odds})$')
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-5, 5)
    ax.set_ylim(0, 1)   
    
    plt.suptitle('Odds, Log-Odds, and Logistic\n', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Print key values
    print("\nKey Values at Different Probabilities:")
    print("-" * 50)
    print(f"{'p':<10} {'Odds':<15} {'Log-Odds':<15}")
    print("-" * 50)
    for p_test in [0.1, 0.25, 0.5, 0.75, 0.9]:
        odds_test = p_test / (1 - p_test)
        log_odds_test = np.log(odds_test)
        print(f"{p_test:<10.2f} {odds_test:<15.3f} {log_odds_test:<15.3f}")

demonstrate_log_odds()
```
