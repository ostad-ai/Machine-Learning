### Machine learning: Gini impurity

The **gini impurity** is a fundamental concept used in decision tree algorithms to measure how *impure* or *mixed* a set of items is with respect to their class labels.
<br>Imagine you have a basket containing different fruits. Gini Impurity measures the probability that you would be wrong if you randomly picked a fruit and guessed its type based on the distribution of fruits in the basket.
<br>**Definition:** Consider we have a dataset of $n$ items having $K$ different types (classes). Let's denote the probability of selecting an item of class $i$ by $p_i$. Then, the Gini impurity is:
<div style="margin-top:4px"></div>
$\large Gini_{impurity}=1-\sum_{i=1}^K p_i^2$
<div style="margin-bottom:4px"></div>

**Hint 1, Lowest value:** When all the items are of the same class, then $Gini_{impurity}=0$, which is the lowest value. As the diversity increases, the Gini impurity also increases.
<br>**Hint 2, Highest value:** For a dataset of $K$ classes, the maximum Gini impurity occurs when all the classea have equal items in the dataset. Thus, $gini_{impurity}=1-\frac {1}{K}$

```python
# Import the required module
from matplotlib import pyplot as plt
```

```python
# Calculate the Gini Impurity for a list of data labels
def gini_impurity(data):
    if len(data) == 0:
        return 0
    
    # Count the occurrences of each class
    classes = set(data) # The classes
    total = len(data) # number of all items
    gini = 1.0
    for c in classes:
        p=data.count(c)/total
        gini -= p ** 2
        
    return gini 
```

```python
# Some examples
# A sample dataset
dataset = ['apple'] * 5 + ['orange'] * 3 + ['banana'] * 4
# Calculate Gini Impurity
impurity = gini_impurity(dataset)
print(f'{dataset}')
print(f"its Gini Impurity is: {impurity:.4f}")

# Perfectly pure case (all same class)
one_class_case = ['apple'] * 12
print('-'*50+f'\n{one_class_case}')
print(f"its Gini impurity is: {gini_impurity(one_class_case):.4f}")

# Worst case (equal distribution)
equal_probs_case = ['apple'] * 4 + ['orange'] * 4+['banana']*4
print('-'*50+f'\n{equal_probs_case}')
print(f"its Gini impurity is: {gini_impurity(equal_probs_case):.4f}")
```

```python
# Draw Gini impurity for a dataset of two classes: A and B
# X-axis the proportion of class A
# Y-axis the value of Gini impurity
ps=[i/100 for i in range(0,100)]
ginis=[1-p**2-(1-p)**2 for p in ps]
plt.plot(ps,ginis,'-')
plt.xlabel('Prob. of  class A')
plt.ylabel('Gini impurity')
plt.title('Gini impurity for a two-class dataset')
plt.show()
```
