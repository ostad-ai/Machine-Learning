#### Logistic function

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# definition of logistic function
def logistic(x,k=1):
    return 1/(1+np.exp(-k*x))
```

```python
xs=np.linspace(-10,10,200)
ks=[.1,.5,1,2,10]
for k in ks:
    ys=logistic(xs,k)
    plt.plot(xs,ys,label=f'k={k}',lw=2.5)
plt.title(r'$f(x)=\frac{1}{1+e^{-kx}}$',fontsize=18)
plt.legend(); plt.show()
```
