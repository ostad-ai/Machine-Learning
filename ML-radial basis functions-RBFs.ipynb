### Radial basis functions (RBFs)
A **radial basis function** (RBF) is a **radial function** which is **positive definite**. It means that the **interpolation** matrix is invertible (nonsingular). A radial function $\phi$ is only dependent on the distance from a center point $\boldsymbol{c}$ and not on the direction:
<br>$\large \phi(\boldsymbol{x},\boldsymbol{c})=\phi(r)$ where $r=||\boldsymbol{x}-\boldsymbol{c}||$
<br> The radial basis function $\phi$ is **stricly positive definite** if for any set of distinct numbers $\boldsymbol{x}_1, \boldsymbol{x}_2,...,\boldsymbol{x}_N$; the interpolation matrix defined below is **nonsingular** (all its **eigenvalues** are positive):
<br>$\phi_{ji}=\phi(||\boldsymbol{x}_j-\boldsymbol{x}_i||)$
<br> In the following we display three strictly positive definite RBFs.

```python
# Import the required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define strictly PD RBFs
# The epsilon > 0 in these functions are called the shape parameter
def gaussian(r, eps=1.0):
    return np.exp(-(eps * r)**2)

def inverse_quadric(r, eps=1.0):
    return 1 / (1 + (eps * r)**2)

def matern32(r, eps=1.0):
    return (1 + np.sqrt(3) * eps * r) * np.exp(-np.sqrt(3) * eps * r)
```

```python
# Display the three RBFs

# Generate distance values
r = np.linspace(0, 5, 500)  # Distance from center

# Compute RBF values
phi_gaussian = gaussian(r)
phi_inverse_quadric = inverse_quadric(r)
phi_matern32 = matern32(r)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(r, phi_gaussian, label="Gaussian: $e^{-(\\epsilon r)^2}$", linewidth=3)
plt.plot(r, phi_inverse_quadric, label="Inverse Quadric: $1/(1 + (\\epsilon r)^2)$", linestyle="--")
plt.plot(r, phi_matern32, label="Matérn (ν=3/2): $(1 + \sqrt{3}\\epsilon r)e^{-\sqrt{3}\\epsilon r}$", linestyle="-.")

# Customize plot
plt.title("Strictly Positive Definite (PD) Radial Basis Functions", fontsize=14)
plt.xlabel("Distance from center ($r$)", fontsize=12)
plt.ylabel("$\\phi(r)$", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(-0.1, 1.1)  # Adjust y-axis limits

# Highlight key properties
plt.annotate(
    "All functions are positive definite\n(guarantee invertible interpolation)",
    xy=(2.5, 0.5), xytext=(3, 0.7),
    arrowprops=dict(facecolor='black', shrink=0.05),
    fontsize=10,
)

plt.show()
```
