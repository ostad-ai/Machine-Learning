In this notebook, we express about Continuous Random Variables with an example from *Continuous uniform distribution*.

```python
# importing the required modules
from matplotlib import pyplot as plt
import numpy as np
```

```python
# continuous uniform distribution with paramters a and b:
def CUD(x,a=0,b=1):
    if x<a:
        return 0
    elif x>b:
        return 1
    else:
        return (x-a)/(b-a)
    
#displaying the distribution function  
a,b=0,1
xs=list(np.linspace(a-1,b+1,100))
ys=[CUD(x,a,b) for x in xs]
plt.plot(xs,ys)
plt.title(f'Continuous Uniform Distribution with parameters $a=${a}, $b=${b}')
plt.show()
```

```python
# probability density fucntion of continuous uniform distribution
# with parameters a and b
def CUP(x,a=0,b=1):
    if x<a or x>b:
        return 0
    else:
        return 1/(b-a)

#display the probability function
a,b=0,1
xs=list(np.linspace(a-1,b+1,100))
ys=[CUP(x,a,b) for x in xs]
plt.plot(xs,ys)
plt.title(f'Probability (Density) Function of Uniform Distr. with parameters $a=${a}, $b=${b}')
plt.show()
```

**Example:** Find the probability $P(0.5<X<2)$ if random variable $X$ comes 
from a uniform distribution with parameters $a=0$ and $b=3$.  
We know that $P(.5<X<2)=F(2)-F(0.5)$. So, let's use the Python function **CUD** defined above to compute it: 

```python
# Solving the example mentioned above:
p_e=CUD(2,a=0,b=3)-CUD(0.5,a=0,b=3)
print('The probability that X falls within 0.5 and 2 for Uniform Distri. with a=0 and b=3:')
print(f'P(0.5<X<2)={p_e}')
```
