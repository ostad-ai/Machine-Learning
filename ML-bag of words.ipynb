# Machine Learning
### Bag fo words
The **Bag of words** (**BoW**) model is a text representation technique used in **natural language processing** (NLP) and **machine learning**. Here, text is converted to a numerical vector in which the **word order** is ignored and only **word frequency** is considered.
<br>Thus, the key idea is to treat each document (e.g., an email, review, sentence) as a **bag** — a collection of words, where:
- Order doesn’t matter
- Only word frequency matters

With BoW, we get a vector representation for each document in a **corpus**. Remember that the word counting is done for each document separately.
<br>**Reminder:** A **corpus** is a collection of texts (or documents) that are gathered for analysis, study, or research.
<hr>

To get BoW for each document in a corpus, we do:
- **Tokenize** the text → split into words.
- Create a **vocabulary** → list of unique words in the entire corpus.
- **Count** occurrences → for each document, count how many times each word appears.
- Represent as **vector** → one entry per word, value = frequency.

<hr>

**Example:** Consider we have only one document in our corpus
- Document: “I love cats. Cats are fun.”
- Then, the vocabulary will be: ["I", "love", "cats", "are", "fun"]
- Finally, the word frequency of the document will be: ["I": 1, "love": 1, "cats": 2, "are": 1, "fun": 1], which means the Bag of Words (BoW) vector is:
  - [1, 1, 2, 1, 1]

<hr>

**Reminder:** What we have defineed above in which we use the frequency of words for the relevant feature values is called **count BoW**. In contrast, if we only consider the presence of word as value 1 and its absence as 0, we get to **binary BoW**, which is suitable for **Bernoulli naive Bayes**.

<hr>

Some **applications** of Bag of Words:
- Text classification (spam detection, sentiment analysis)
- Information retrieval

<hr>

In the following:
- We implement a `preprocess_text` function to tokenize the texts
- We also implement a `build_vocabulary` function to build the vocabulary given the entire corpus.
- After that, we implement a `BagOfWords` class to vectorize a corpus. 
- Finally, we test our BoW with a synthetic corpus.

```python
# Import required modules
import numpy as np
import re
```

```python
# ================================
# 1. Preprocessing: Tokenize and Clean Text
# ================================

def preprocess_text(text):
    """
    Clean and tokenize text.
    """
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and extra whitespace
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    # Split into words
    words = text.split()
    # Remove empty strings
    words = [word for word in words if word]
    return words

# ================================
# 2. Build Vocabulary
# ================================

def build_vocabulary(texts):
    """
    Build a set of unique words from all texts.
    """
    vocab = set()
    for text in texts:
        words = preprocess_text(text)
        vocab.update(words)
    return sorted(vocab)

# ================================
# 3. Bag of Words Vectorizer
# ================================

class BagOfWords:
    def __init__(self):
        self.vocabulary = []
        self.word_to_idx = {}
        self.idx_to_word = {}

    def fit(self, texts):
        """
        Build vocabulary from texts.
        """
        self.vocabulary = build_vocabulary(texts)
        self.word_to_idx = {word: i for i, word in enumerate(self.vocabulary)}
        self.idx_to_word = {i: word for i, word in enumerate(self.vocabulary)}
        return self

    def transform(self, texts):
        """
        Convert texts to BoW vectors.
        """
        vectors = []
        for text in texts:
            words = preprocess_text(text)
            vector = np.zeros(len(self.vocabulary))
            for word in words:
                if word in self.word_to_idx:
                    vector[self.word_to_idx[word]] += 1
            vectors.append(vector)
        return np.array(vectors)

    def fit_transform(self, texts):
        """
        Fit and transform in one step.
        """
        return self.fit(texts).transform(texts)

    def get_vocabulary(self):
        return self.vocabulary

    def get_word_to_idx(self):
        return self.word_to_idx

    def get_idx_to_word(self):
        return self.idx_to_word
```

```python
# ================================
# 4. Example Usage (Clean)
# ================================

# Sample texts — cleaned and simple
texts = [
    "I love cats and dogs",
    "Cats are fun, dogs are loyal",
    "I love cats more than dogs",
    "Dogs are great pets",
    "I love cats and dogs again"
]

# Step 1: Fit BoW model
bow = BagOfWords()
bow.fit(texts)

# Step 2: Get vocabulary
print("Vocabulary:", bow.get_vocabulary())

# Step 3: Transform to vectors
vectors = bow.transform(texts)
print(f'Shape of vectors:{vectors.shape}; for {vectors.shape[0]} documents,'+\
      f'and vocabulary size: {vectors.shape[1]}')
print("Sample vector:", vectors[1])

# Step 4: Show mapping
print('-'*30)
print("Word to Index Mapping:")
for word, idx in bow.get_word_to_idx().items():
    print(f"{word} → {idx}", end=', ')
print('\n'+'-'*30)
print("Index to Word Mapping:")
for idx, word in bow.get_idx_to_word().items():
    print(f"{idx} → {word}",end=', ')
```
