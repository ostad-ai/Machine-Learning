## Gradient and Tangent Planes
Given a hypersurface by $f(\boldsymbol{x})=constant$, its gradient $\nabla f$ is orthogonal to the hypersurface. Thus, the tangent hyperplane at point $\boldsymbol{x}_0$ is obtained by: <br>
$\nabla f(\boldsymbol{x}_0)\cdot (\boldsymbol{x}-\boldsymbol{x}_0)=0$

```python
# importing required modules
import numpy as np
from matplotlib import pyplot as plt
```

The level curve for $f(x,y)=x^2+y^2$ defined by $f(x,y)=r^2$ is given.
We want to find the tangent line to the point of interest, $(x_0,y_0)$, on the curve.
<br> For this purpose, we take the gradient of $f$:
<br>$\nabla f(x_0,y_0)=[2x_0,2y_0]^T$
<br> Now, we put the gradient into the equation for hyperplanes:
<br> $[2x_0,2y_0]^T\cdot[x-x_0,y-y_0]^T$=0
<br> simplifiyng above equation leads to:
<br> $x_0(x-x_0)+y_0(y-y_0)=0 \rightarrow xx_0+yy_0=x^2_0+y^2_0=r^2$
<br> Finally, we get the tangent line:
<br>$xx_0+yy_0=r^2$

```python
# tangent line to the level curve x^2+y^2=r^2
teta=np.linspace(0,2*np.pi,100)
# radius of a circle and angle of the point
for r,alpha in [(9,3.9),(15,.6),(25,1.9)]:
    # xs and ys are the points on the circle
    xs=r*np.cos(teta)
    ys=r*np.sin(teta)
    # the point of interest with angle and radius
    # x_0=r*cos(alpha), y_0=r*sin(alpha)
    #--tangent line
    # x*cos(alpha)+y*sin(alpha)=r
    # xt and yt are the points on the tangent line
    xt=np.linspace(-1.3*r,1.3*r,100)
    yt=(r-xt*np.cos(alpha))/np.sin(alpha)
    plt.plot(xs,ys,'-'); plt.plot(xt,yt,'-')
    plt.plot(r*np.cos(alpha),r*np.sin(alpha),'*')
plt.axis('equal')
plt.title('Tangent lines to the level curves: $x^2+y^2=r^2$')
plt.show()
```
