```python
from matplotlib import pyplot as plt
from sklearn import datasets
import numpy as np
```

```python
# loading the dataset
iris=datasets.load_iris()
data=iris.data
feature0=data[:,0] #a column of dataset
```

```python
# showing elements of boxplot for one column (feature) of dataset
maxf,minf,medianf=feature0.max(),feature0.min(),np.median(feature0)
mediantf,medianbf=np.median(feature0[np.where(feature0>medianf)]),\
np.median(feature0[np.where(feature0<medianf)])
plt.boxplot(feature0)
plt.text(1.05,maxf,f'maximum:{maxf}')
plt.text(1.05,minf,f'minimum:{minf}')
plt.text(1.1,medianf,f'median:{medianf}')
plt.text(1.1,mediantf,f'third quartile:{mediantf}')
plt.text(1.1,medianbf,f'first quartile:{medianbf}')
plt.suptitle(f'Boxplot elements')
plt.title(f'IQR=$third\; quartile-first\; quartile$={mediantf-medianbf:.2f}')
plt.show()
```

```python
# plotting all features of the dataset with boxplots
# each column of the dataset is a feature
# small dots are outliers
plt.boxplot(data)
plt.show()
```
